```python
import pandas as pd
import numpy as np
```

```python
from pandasql import sqldf
```

## Operations on the Dataframe

```python
df=pd.read_excel("dfemlakjet1.xlsx")
df.head()
```

```python
df["Wc Sayısı"].value_counts()
```

```python
query=""" select * from df where Semt like"%Satılık Bina%" or Semt like"%Satılık Yalı%" or Semt like"%Satılık Köşk%"  """
drop_rows=sqldf(query)
drop_rows
```

```python
df2=pd.concat([df,drop_rows]).drop_duplicates(keep=False)
df2.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
```

```python
df2=df2.dropna()
df2['index']=np.arange(len(df2))
df2=df2.set_index("index")
df2.info()
```

```python
df2["Wc Sayısı"].isnull().sum()
```

```python
df2["Semt"].value_counts()
```

```python
df2["Wc Sayısı"].value_counts()
```

```python
df2["Binanın Kat Sayısı"] = df2['Binanın Kat Sayısı'].astype('int64')
df2["Banyo Sayısı"] = df2['Banyo Sayısı'].astype('int64')
df2["Fiyat"] = df2['Fiyat'].astype('int64')
df2["Brüt Metrekare"] = df2['Brüt Metrekare'].astype('int64')
```
