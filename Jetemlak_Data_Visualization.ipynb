## Removing Outliers from Data

```python
Q1 = np.percentile(df2['Fiyat'], 25,
                   interpolation = 'midpoint')
Q3 = np.percentile(df2['Fiyat'], 75,
                   interpolation = 'midpoint')
IQR=Q3-Q1

print("Outlierlar silinmeden önce veri sayısı: ", df2.shape)
z=len(df2)
```

```python
upper = np.where(df2['Fiyat'] >= (Q3+1.5*IQR))
df2=df2.drop(upper[0],axis=0)

lower = np.where(df2['Fiyat'] <= (Q1-1.5*IQR))
df2=df2.drop(lower[0],axis=0)
```

```python
df2=df2.reset_index(drop=True)
df2.drop(["index"],axis=1, inplace=True)
y=len(df2)
print("Outlierlar silindikten sonra veri sayısı:",df2.shape,"\nSilinen Outlier Sayısı:",z-y)
```
